```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city=pd.read_csv('/content/dim_cities.csv')
respondents=pd.read_csv('/content/dim_repondents.csv')
survey=pd.read_csv('/content/fact_survey_responses.csv')
```

```python
city.head()
```

```python
respondents.head()
```

```python
a=pd.merge(city,respondents).drop(['Tier'],axis=1)
a.head()
```

```python
survey.head()
```

```python
survey1=pd.merge(survey,a)
survey1.head()
```

```python
survey1.columns
```

```python
survey1[['Reasons_preventing_trying', 'Current_brands',
       'Reasons_for_choosing_brands', 'Improvements_desired',
       'Ingredients_expected', 'Health_concerns',
       'Interest_in_natural_or_organic']]
```

```python
b=survey1[['Marketing_channels','Age']].groupby('Marketing_channels').value_counts().reset_index()
b
```

# ***Demographic Insights***

```python
plt.figure(figsize=(10,5))
sns.heatmap(b.pivot(index='Marketing_channels',columns='Age',values='count'),cmap='Blues',annot=True)
plt.title('Marketing Channels')
plt.show()
```

```python
#Who prefers energy drink more?
plt.pie(survey1.Gender.value_counts(),labels=['Male','Female','Non-binary'],autopct='%1.1f%%')
plt.title('Who prefers energy drink more? ')
plt.show()
```

```python
#Which age group prefers energy drinks more?
sns.countplot(data=survey1,x='Age',palette='hls')
plt.grid()
plt.title('Which age group prefers energy drinks more?')
plt.show()
```

```python
#Respndoents by age group and gender
sns.countplot(data=survey1,x='Age',hue='Gender')
plt.title('Respndoents by age group and gender')
plt.grid()
plt.show()
```

# ***Consumer Preferences***

```python
#Consemption frequency by Gender
plt.figure(figsize=(10,4))
sns.countplot(data=survey1,x='Consume_frequency',hue='Gender',palette='husl')
plt.grid()
plt.title('Consemption frequency by Gender')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(14,5))
fig.suptitle('Consumer Preferences')
sns.countplot(data=survey1,x='Ingredients_expected',palette='Set2',ax=ax[0])
sns.countplot(data=survey1,x='Packaging_preference',palette='Paired',ax=ax[1])
plt.show()
```

***Competition Analysis***

```python
fig,ax=plt.subplots(1,2,figsize=(14,4))
fig.suptitle('Competition Analysis')
sns.countplot(data=survey1,x='Reasons_for_choosing_brands',palette='mako',ax=ax[0])
sns.countplot(data=survey1,x='Current_brands',palette='rocket',ax=ax[1])
plt.show()
```

# ***Purchase Behavior***

```python
plt.pie(survey1.Price_range.value_counts(),labels=['50-99','100-150','Above 150','Below 50'],autopct='%1.1f%%')
plt.title('Price Preference')
plt.show()
```

```python
plt.pie(survey1.Purchase_location.value_counts(),labels=['Supermarkets','Online retailers','Gyms and fitness centers','Local stores','Other'],autopct='%1.1f%%')
plt.title('Purchasing Locations')
plt.show()
```

```python
##Consumption situations by age groups
ch=survey1[['Typical_consumption_situations','Age']].groupby('Typical_consumption_situations').value_counts().reset_index()

ch.pivot(index='Typical_consumption_situations',columns='Age',values='count')
```

```python
##Reasons_for_choosing_brands
ss=survey1[['Reasons_for_choosing_brands','Gender']].groupby('Reasons_for_choosing_brands').value_counts().reset_index()
ss.pivot(index='Reasons_for_choosing_brands',columns='Gender',values='count')
```

```python
##Reasons_preventing_trying
r=survey1[['Reasons_preventing_trying','Age']].groupby('Reasons_preventing_trying').value_counts().reset_index()
r.pivot(index='Reasons_preventing_trying',columns='Age',values='count')
```

```python
##General Perception
bb=survey1[['General_perception','Current_brands']].groupby('Current_brands').value_counts().reset_index()
bb['count']=bb['count']/bb['count'].sum()*100
bb['count']=bb['count'].round(2)
bb['count']=bb['count'].map('{:.2f}%'.format)
bb.pivot(index='Current_brands',columns='General_perception',values='count')
```
